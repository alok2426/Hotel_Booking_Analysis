## 3. ***Data Wrangling***

### Cleaning data

```python
null_columns = ['company','agent','children']
for column in null_columns:
  df_new[column].fillna(0,inplace=True)

null_columns2=['country']
for col in null_columns2:
  df_new[null_columns2].fillna('other',inplace=True)
```

```python
# let's check

df_new.isna().sum().sort_values(ascending=True)[:10].reset_index().rename(columns={'index':'columns',0:'Null values'})
```

```python
df_new['total_people'] = df_new['adults'] + df_new['babies'] + df_new['children']
df_new['total_stay'] = df_new['stays_in_weekend_nights'] + df_new['stays_in_week_nights']
```

## ***4. Data Vizualization & Experimenting with charts : Understand the relationships between variables***



```python
df_new['hotel'].value_counts().reset_index()
```

```python
# Chart - 1 visualization code
df_new['hotel'].value_counts().plot.bar(figsize=(10,5),fontsize=15)
plt.xlabel('Type of hotel',color='green',size='20')
plt.ylabel('Number of bookings',{'size':20},color='green')
plt.show()
```

```python
# booking in terms of percentage via pie chart
df_new['hotel'].value_counts().plot.pie(explode=(0.03,0.03),shadow=True,figsize=(10,5),fontsize=20,autopct='%1.1f%%')
plt.title('Most booked hotel',size='20',color='green')
```

```python
# highest bookings made by agents.
max_bookings=df_new.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count':'No. of bookings'}).sort_values(by='No. of bookings',ascending=False)
# thier is agent with zero number,so lets drop that agent.
max_bookings.drop(max_bookings[max_bookings['agent']==0].index,inplace=True)
top_ten_agents=max_bookings[:10]
top_ten_agents
```

```python
# Chart - 2 visualization code
plt.figure(figsize=(12,6))
sns.barplot(x=top_ten_agents['agent'],y=top_ten_agents['No. of bookings'])
plt.title('Booking counts by agents',size='20',color='red')
plt.xlabel('Agent number',size=18)
plt.ylabel('No of bookings',size=18)
```

#### Chart - 8
##8) Let's check which month got maximum bookings

```python
# Chart - 8 - code
monthly_bookings = df_new.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':'counts'})
#to put months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_bookings['arrival_date_month']=pd.Categorical(monthly_bookings['arrival_date_month'],categories=months,ordered=True)
monthly_bookings=monthly_bookings.sort_values('arrival_date_month')
monthly_bookings
```

```python
# visualization code
plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_bookings['arrival_date_month'],y=monthly_bookings['counts'])

plt.title('Number of bookings in each month',color='blue',size=20)
plt.xlabel('Month',size=20)
plt.ylabel('Number of bookings',size=25)
```

#### Chart - 10 - Correlation Heatmap

```python
#Chart - 10 - Correlation Heatmap
# Correlation Heatmap visualization code
# Convert categorical columns to numerical using one-hot encoding
# Select only numerical features for correlation analysis
numerical_features = df_new.select_dtypes(include=np.number)

plt.figure(figsize=(16, 8))
sns.heatmap(numerical_features.corr(), annot=True)
plt.title('Co-relation of the columns', size=25, color='blue')
plt.show()
```
